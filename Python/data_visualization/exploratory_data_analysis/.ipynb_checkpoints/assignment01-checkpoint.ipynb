### Corelations

```python
import seaborn as sns
# styling
sns.set_style("darkgrid")

palette = sns.color_palette("hls", 10)
sns.set_palette(palette)
# bivariate descriptive statistics for two categorical variables
def bivariate_cat_cat_ds(intakes, categorical1, categorical2):
    frequency_table = pd.crosstab(intakes[categorical1], intakes[categorical2])
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
    
    sns.heatmap(frequency_table, cmap='Blues', annot=True, fmt='g', ax=ax1)
    sns.heatmap(frequency_table.div(len(intakes)), cmap='Blues', annot=True, fmt='g', ax=ax2)
    
bivariate_cat_cat_ds(intakes, 'Animal Type', 'Intake Condition')
```
