Z datasetu vybereme pouze *object* sloupce a ty rozdělíme podle počtu unikátních hodnot na **high_uniq** a **low_uniq**.
* **high_uniq** - jména sloupců, co mají více než 3 unikátní hodnoty
* **low_uniq** - jména sloupců, co mají 3 nebo méně unikátních hodnot

```python
# Get columns with 3 or less unique records
low_uniq = []
# Get columns with more than 3 unique records
high_uniq = []
# Get columns which are only 'object' datatype
objects = []

for dt in data.dtypes.items():
    if dt[1] == object:
        objects.append(dt[0])

for i in objects:
    if len(data[i].unique()) <= 3:
        low_uniq.append(i)
    if len(data[i].unique()) > 3:
        high_uniq.append(i)
print("('object') Columns with low unique records: ")
display(low_uniq)
print("('object') Columns with high unique records: ")
display(high_uniq)
```

U všech příznaků z **low_uniq**, které mají pouze dvě možné hodnoty, změníme tyto hodnoty na 1 nebo 0.
U všech příznaků z **low_uniq**, které mají pouze tři možné hodnoty, tyto hodnoty přeměníme buď na 2 nebo 1 nebo 0.
Dále je přetypujeme na datový typ int.

```python
# Change 2 unique values to 1 and 0
def changeToBinary ( column, one, zero):
    data[column] = data[column].replace([one,zero], [1,0])
    data[column] = data[column].astype(int)

changeToBinary( 'Gender', 'M', 'F')
changeToBinary( 'ParentAnsweringSurvey', 'Yes', 'No')
changeToBinary( 'ParentschoolSatisfaction', 'Good', 'Bad')
changeToBinary( 'StudentAbsenceDays', 'Above-7', 'Under-7' )
changeToBinary( 'Relation', 'Father', 'Mum' )
changeToBinary( 'Semester', 'S', 'F' )

# Change 3 unique values to 2, 1 and 0 
def changeToTrinary ( column, two, one, zero):
    data[column] = data[column].replace([two,one,zero], [2,1,0])
    data[column] = data[column].astype(int)
    
changeToTrinary('StageID', 'lowerlevel', 'MiddleSchool', 'HighSchool')
changeToTrinary('SectionID', 'A', 'B', 'C')
changeToTrinary('Class', 'M', 'L', 'H')
```

Jelikož si rozhodovací strom neumí poradit s nenumerickými příznaky, typ všech těchto příznaků nejdříve převedeme na "category" a následně na číselnou reprezentaci pomocí cat.codes.

# Rozdělení datasetu na trénovací, validační a testovací části

Data rozdělíme na *trénovací*, *testovací* a *validační* skupinu tak, aby 60% dat připadlo trénovací skupině, 20% validační a 20% testovací.
Data se rozdělí pomocí funkce **train_test_split**, která to provede pseudonáhodně.

```python
random_seed = 333

# Get rid of explained variable in Xdatasets and save explained variable to ydatasets
Xdata = data.drop("ParentschoolSatisfaction", axis = 1)
ydata = data.ParentschoolSatisfaction

# Split data: 60% for train data, 40% rest
Xtrain, Xrest, ytrain, yrest = train_test_split(Xdata, ydata, test_size=0.4, random_state=random_seed) 
print("Xtrain: ", Xtrain.shape, "\nytrain: ", ytrain.shape, "\nXtest: ", Xrest.shape, "\nytest: ", yrest.shape, "\n")

# Split rest of data (40%) into valid and test data, 50%/50%
Xval, Xtest, yval, ytest = train_test_split(Xrest, yrest, test_size=0.5, random_state=random_seed) 
print("Xval: ", Xval.shape, "\nyval: ", yval.shape, "\nXtest: ", Xtest.shape, "\nytest: ", ytest.shape)
```

# Rozhodovací strom
### Najdeme nejlepší hyperparametry a naučíme model

* Najdeme nejlepší hodnoty pro tyto hyperparametry:
    * Depth = {1, 50}
    * Criterion = {entropy, gini, log_loss}
    * min_samples_leaf = {1, 4}
* Naučíme model
    * Použijeme nejlepší parametery a použijeme **.fit** na trénovací dataset
* Zobrazíme výsledek
    * Graf
    * Přesnost - pomocí metody metrics.accuracy_score


```python
# Find best hyperparameters (depths, criterions, min_samples_leaf)
get_param = {'max_depth': range(1,51),
             'criterion': ['entropy', 'gini', 'log_loss'],
             'min_samples_leaf': range(1,4)}
param_comb = ParameterGrid(get_param)

# Try all combinations (depth * criterions)
valid_acc = []
train_acc = []
for get_param in param_comb:
    dec_tree = DecisionTreeClassifier(max_depth=get_param['max_depth'],
                                      criterion=get_param['criterion'],
                                      min_samples_leaf=get_param['min_samples_leaf'])
    dec_tree.fit(Xtrain, ytrain)
    train_acc.append(metrics.accuracy_score(ytrain, dec_tree.predict(Xtrain)))
    valid_acc.append(metrics.accuracy_score(yval, dec_tree.predict(Xval)))

# Get best parameters from function above
best_param = param_comb[np.argmax(valid_acc)]
print("Best_params:", best_param, "\n")

# Train our model with best hyperparameters
dec_tree = DecisionTreeClassifier(**best_param)
dec_tree.fit(Xtrain, ytrain)

# Save our test prediction for later comparisons
tree_pred = metrics.accuracy_score(ytest, dec_tree.predict(Xtest))

# Plot grapth
fig, ax = plt.subplots(figsize=(16,5))
ax.plot(train_acc, 'o-', color='purple')
ax.plot(valid_acc, 'o-', color='blue')
ax.grid(axis='y', color='black', alpha=.5, linewidth=.5)
ax.set_title('Accuracy score with decision tree')
plt.xlabel('Hyperparameter index')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='center right')


# Print our accuracies
print('Accuracy score (train): {0:.4f}'.format(metrics.accuracy_score(ytrain, dec_tree.predict(Xtrain))))
print('Accuracy score (validation): {0:.4f}'.format(metrics.accuracy_score(yval, dec_tree.predict(Xval))))
print('Accuracy score (test): {0:.4f}'.format(metrics.accuracy_score(ytest, dec_tree.predict(Xtest))))
```

# K-nejbližších sousedů
### Najdeme nejlepší hyperparametry a naučíme model

* Najdeme nejlepší hodnoty pro tyto hyperparametry:
    * n_neighbors = {3, 15}
* Naučíme model
    * Použijeme nejlepší parametery a použijeme **.fit** na trénovací dataset
* Zobrazíme výsledek
    * Graf
    * Přesnost - pomocí metody metrics.accuracy_score

```python
val_acc = []
train_acc = []

kneighbors = range(3,15)
for k in kneighbors:
    clf = KNeighborsClassifier(n_neighbors = k)
    clf.fit(Xtrain, ytrain)
    val_acc.append(clf.score(Xval, yval))
    train_acc.append(clf.score(Xtrain, ytrain))

# Plot grapth
plt.figure(figsize=(20,6))
plt.plot(kneighbors, train_acc,'o-', color='purple') #o: body, r: barva, -: spojení bodů linkou
plt.plot(kneighbors, val_acc,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])

best_params_knn = np.argmax(val_acc) + 3
print('Best number of neighbours: ', best_params_knn, '\n')
best_val_acc_knn = val_acc[np.argmax(val_acc)]

clf = KNeighborsClassifier(n_neighbors = best_params_knn)
clf.fit(Xtrain, ytrain)

# Print our accuracies
print('Accuracy score (train): {0:.4f}'.format(metrics.accuracy_score(ytrain, clf.predict(Xtrain))))
print('Accuracy score (validation): {0:.4f}'.format(metrics.accuracy_score(yval, clf.predict(Xval))))
print('Accuracy score (test): {0:.4f}'.format(metrics.accuracy_score(ytest, clf.predict(Xtest))))
```

Z grafu můžeme vypozorovat jakou přesnost dávají různé počty sousedů.

# Výběr nejlepšího modelu
* Na závěr porovnáme přesnost **rozhodovacího stromu** na **validační** množině s přesností **KNN** na **validační** množině.
* Vybereme lepší přesnost.
* Zobrazíme finální přesnost na **testovaích** datech.

```python
print('Accuracy score of decision tree (validation): {0:.4f}'.format(metrics.accuracy_score(yval, dec_tree.predict(Xval))))
print('Accuracy score of KNN (validation): {0:.4f}'.format(metrics.accuracy_score(yval, clf.predict(Xval))))
print("\n----------------------------------------------------\n")

dc = metrics.accuracy_score(yval, dec_tree.predict(Xval))
knn = metrics.accuracy_score(yval, clf.predict(Xval))

# from float to 'decimal' for easier comparison
a = round(dc*100, 0)
b = round(knn*100, 0)

if a >= b:
    print("Decision tree has better accuracy on validation dataset!")
    print(f"Accuracy score (test): {100*dc:.2f} %")
if b > a:
    print("KNN has better accuracy on validation dataset!")
    print(f"Accuracy score (test): {100*knn:.2f} %")
```

Příprava dat pro vykreslení distribuce chyby.

```python
def create_error_df(yreal, ypred, subset):
    df = pd.DataFrame()
    df['real'] = yreal
    df['predicted'] = ypred
    df['error'] = yreal - ypred
    df['set'] = subset
    return df
    
df = pd.concat([
    create_error_df(ytrain, dec_tree.predict(Xtrain), 'train'),
    create_error_df(yval, dec_tree.predict(Xval), 'validation'),
    create_error_df(ytest,dec_tree.predict(Xtest), 'test'),
])

display(df.head())
df.describe()
```

Pomocí **violin plotu** si vykreslíme distribuci chyby.

```python
# seaborn default theme with dark grid
sns.set_theme()

fig, ax = plt.subplots(figsize=(16,8))
sns.violinplot(ax=ax, data=df, x='set', y='error', palette='Set2')

ax.set_ylabel('Error')
ax.set_xlabel('Set')
_ = ax.set_title('Error across sets')
```
